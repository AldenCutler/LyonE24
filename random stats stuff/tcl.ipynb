```python
import folium
import geopandas as gpd
```

```python
data = gpd.read_file('tcl_lines.json')
```

```python
date_debut = data['date_debut'] 
date_fin = data['date_fin']
last_update = data['last_update']
last_update_fme = data['last_update_fme']
data = data.drop(columns=['date_debut', 'date_fin', 'last_update', 'last_update_fme'])
data
```

```python
colors = []
for index, row in data.iterrows():
    if row['ligne'] == 'D': # green line
        colors.append('#029C41')
    elif row['ligne'] == 'A':   # pink line
        colors.append('#E62E86')
    elif row['ligne'] == 'B':   # blue line
        colors.append('#0065B1')
    elif row['ligne'] == 'C':   # yellow line
        colors.append('#F48E06')
    else:   # furnicular line, light green
        colors.append("#93BF38")
        
data['style'] = [
    {
        "color": colors[i],
        "weight": 4,
        "opacity": 1
    }
    for i in range(len(data))
]
```

```python
m = folium.Map(location=[45.75, 4.85], zoom_start=13)

folium.GeoJson(data, style_function=lambda x: x['properties']['style']).add_to(m)

m.save('tcl_lines.html')
```
